```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
#implementing principle component analysis
from sklearn.decomposition import PCA
pca_30 = PCA(n_components = 10,random_state = 2020)
pca_30.fit(X_scaled)
X_pca_30 = pca_30.transform(X_scaled)
```

```python
print("variance by 30 principle components",
     sum(pca_30.explained_variance_ratio_ *100))
```

```python
pca_30.explained_variance_ratio_ *100
```

```python
#cumulative sum
np.cumsum(pca_30.explained_variance_ratio_ *100)
```

```python
plt.plot(np.cumsum(pca_30.explained_variance_ratio_))
plt.xlabel('Number of components')
plt.ylabel('Variance')
```

```python
pca_2 = PCA(n_components = 2,random_state = 2020)
pca_2.fit(X_scaled)
X_pca_2 = pca_2.transform(X_scaled)
```

```python
# Performing clustering
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i,init = 'k-means++',n_init = 12, random_state = 0)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
    plt.show()
```

```python
f3, ax = plt.subplots(figsize = (8, 6))
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 2)
clusters = kmeans.fit_predict(X_scaled)
```

```python
colors = 'rgbkcmy'
for i in np.unique(clusters):
    plt.scatter(X_scaled[clusters == i, 0],
               X_scaled[clusters == i, 1],
               color = colors[i], label = 'cluster' + str(i+1))
plt.legend()
```
